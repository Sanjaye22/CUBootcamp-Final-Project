```python
import pandas as pd
import warnings
warnings.simplefilter('ignore')

# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import string
```

```python
file_test = "finaltest.csv"
file_train = "finaltrain.csv"
```

```python
test_df = pd.read_csv(file_test, encoding="ISO-8859-1")
test_df.head()
```

```python
train_df = pd.read_csv(file_train, encoding="ISO-8859-1")
train_df.head()
```

```python
#Swapping all uppercase to lowercase in Test dataset
test_df["review"]= test_df["review"].str.lower()
test_df.head()
```

```python
#Swapping all uppercase to lowercase in Train dataset
train_df["review"]= train_df["review"].str.lower()
train_df.head()
```

```python
#Removing all punctuation from review column
string.punctuation
```

```python
# Removing "" marks from review column
test_df["review"] = test_df["review"].str.replace('[""]','')

test_df_new= test_df
test_df_new.head()
```

```python
# Removing all punctuations 
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct


test_df_new["review"]=test_df_new["review"].apply(lambda x: remove_punctuation(x))
test_df_new
```

```python
# Removing "" marks from review column
train_df["review"] = train_df["review"].str.replace('[""]','')

train_df_new= train_df
train_df_new.head()
```

```python
# Removing all punctuations 
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct


train_df_new["review"]=train_df_new["review"].apply(lambda x: remove_punctuation(x))
train_df_new
```

```python
test_df_new.review
```

```python
# Removing all numeric values in the review column
test_df_new['review'] = test_df_new['review'].str.replace('\d+', '')
test_df_new.head()
```

```python
# Removing all numeric values in the review column
train_df_new['review'] = train_df_new['review'].str.replace('\d+', '')
train_df_new.head()
```

```python
# Removing all unicodes in review column
test_df_new= test_df_new.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))

test_df_new.head()
```

```python
# Removing all unicodes in review column
train_df_new= train_df_new.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))

train_df_new.head()
```

```python
# Switching rating column values to 0 & 1

test_df_new['rating']=test_df_new['rating'].replace(["1","2","3","4","5"], 0)

test_df_new['rating']=test_df_new['rating'].replace(["6","7","8","9","10"], 1)

test_df_new.head()
```

```python
# Switching rating column values to 0 & 1

train_df_new['rating']=train_df_new['rating'].replace(["1","2","3","4","5"], 0)

train_df_new['rating']=train_df_new['rating'].replace(["6","7","8","9","10"], 1)

train_df_new.head()
```

```python
# Exporting cleaned datasets to csv files
test_df_new.to_csv('testfinal.csv',index=False)
train_df_new.to_csv('trainfinal.csv',index=False)
```
